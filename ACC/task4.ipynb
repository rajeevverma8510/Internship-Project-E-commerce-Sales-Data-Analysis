# **EXPLORATORY DATA ANALYSIS**

## Sales 

```python
grouped_df= df.groupby(['Region'],as_index=False)['Sales'].sum().sort_values(by='Sales',ascending=False)
plt.figure(figsize=(18,15))
bars=plt.bar(grouped_df['Region'], grouped_df['Sales'],color='red')
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height + 0.2, 
             f'{height}', 
             ha='center',va='bottom')

plt.title('Total Sales According in Region')
plt.xlabel('Sales')
plt.ylabel('Total Sales in million in Region')
plt.xticks(rotation=45)
plt.show()
```

## Discount 

```python
grouped_df= df.groupby(['Region'],as_index=False)['Discount'].sum().sort_values(by='Discount',ascending=False)
plt.figure(figsize=(18,15))
bars=plt.bar(grouped_df['Region'], grouped_df['Discount'],color='Blue')
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height + 0.2, 
             f'{height}', 
             ha='center',va='bottom',rotation=90)

plt.title('Total Sales According in Region')
plt.xlabel('Discount')
plt.ylabel('Total Discount in million in Region')
plt.xticks(rotation=45)
plt.show()
```

## Profit

```python
grouped_df= df.groupby(['Region'],as_index=False)['Profit'].sum().sort_values(by='Profit',ascending=False)
plt.figure(figsize=(20,15))
bars=plt.bar(grouped_df['Region'], grouped_df['Profit'],color='Green')
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height + 0.2, 
             f'{height}', 
             ha='center',va='bottom',rotation=90)

plt.title('Total Sales According in Region')
plt.xlabel('Profit')
plt.ylabel('Total Profit in million in Region')
plt.xticks(rotation=45)
plt.show()
```

```python
import folium
import plotly.express as px


# Group data by Region and Country to calculate total sales, profit, and order count
region_sales = df.groupby(['Region', 'Country']).agg({
    'Sales': 'sum',
    'Profit': 'sum',
    'Order ID': 'count',
    'Country latitude': 'first',
    'Country longitude': 'first'
}).reset_index()

# Rename columns for clarity
region_sales.rename(columns={'Order ID': 'Order Count'}, inplace=True)

#  Bar Plot - Total Sales by Region
sns.set_theme(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.barplot(x='Sales', y='Region', data=region_sales, estimator=sum)
plt.title('Total Sales by Region')
plt.show()

#  Bar Plot - Total Profit by Region
plt.figure(figsize=(10, 6))
sns.barplot(x='Profit', y='Region', data=region_sales, estimator=sum)
plt.title('Total Profit by Region')
plt.show()

# Map Visualization with Folium (Markers for countries)
mymap = folium.Map(location=[20, 0], zoom_start=2)  # Create a map centered at a general location

# Add markers for each country
for index, row in region_sales.iterrows():
    lat = row['Country latitude']
    lon = row['Country longitude']
    
    folium.Marker([lat, lon], 
                  popup=f"Country: {row['Country']}<br>Sales: ${row['Sales']:.2f}<br>Profit: ${row['Profit']:.2f}"
                 ).add_to(mymap)

# Save the map as an HTML file to view in a browser
mymap.save('geographic_sales_map.html')

#  Choropleth Map - Sales by Country using Plotly
fig = px.choropleth(region_sales, 
                    locations="Country", 
                    locationmode="country names", 
                    color="Sales", 
                    hover_name="Country", 
                    color_continuous_scale="Blues", 
                    title="Sales by Country")

# Show the interactive map
fig.show()
```

```python
#  Group by Category and Subcategory to calculate total profit and sales
category_profit = df.groupby(['Category', 'Subcategory']).agg({'Profit': 'sum', 'Sales': 'sum'}).reset_index()

#  Sort by profit to see the most and least profitable categories and subcategories
category_profit = category_profit.sort_values(by='Profit', ascending=False)

#  Visualize the profitability by Category and Subcategory

#  Bar Plot for Total Profit by Category
sns.set_theme(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.barplot(x='Profit', y='Category', data=category_profit, estimator=sum)
plt.title('Total Profit by Category')
plt.show()

# 4.2: Bar Plot for Total Profit by Subcategory
plt.figure(figsize=(12, 8))
sns.barplot(x='Profit', y='Subcategory', data=category_profit, estimator=sum)
plt.title('Total Profit by Subcategory')
plt.xticks(rotation=90)
plt.show()

#  Heatmap for Profit by Category and Subcategory

# Pivot the data for a heatmap
category_subcategory_profit = category_profit.pivot_table(index='Category', columns='Subcategory', values='Profit')

# Create a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(category_subcategory_profit, annot=True, cmap="coolwarm", linewidths=.5)
plt.title('Profitability Heatmap by Category and Subcategory')
plt.show()
```

```python

#  Convert 'Order Date' to datetime format
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')

# Check for any issues in the conversion
if df['Order Date'].isnull().any():
    print("There are invalid date formats in the dataset. Please check the 'Order Date' column.")

#  Extract Month and Year for grouping
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month_name()  # This will give us the month name
df['Month-Year'] = df['Order Date'].dt.to_period('M')  # For better grouping

#  Group data by Month-Year to calculate total sales, profit, and order quantity
monthly_trends = df.groupby('Month-Year').agg({
    'Sales': 'sum',
    'Profit': 'sum',
    'Quantity': 'sum'  # Assuming 'Quantity' is the name of the column
}).reset_index()

# Convert Month-Year back to a string for plotting
monthly_trends['Month-Year'] = monthly_trends['Month-Year'].astype(str)

#  Visualize Trends

# Set the Seaborn style
sns.set_theme(style="whitegrid")

# Line Plot for Sales Over Time
plt.figure(figsize=(14, 8))
sns.lineplot(data=monthly_trends, x='Month-Year', y='Sales', marker='o')
plt.title('Monthly Sales Trends')
plt.xlabel('Month-Year')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()

#  Line Plot for Profit Over Time
plt.figure(figsize=(14, 8))
sns.lineplot(data=monthly_trends, x='Month-Year', y='Profit', marker='o', color='orange')
plt.title('Monthly Profit Trends')
plt.xlabel('Month-Year')
plt.ylabel('Total Profit')
plt.xticks(rotation=45)
plt.show()

#  Line Plot for Order Quantity Over Time
plt.figure(figsize=(14, 8))
sns.lineplot(data=monthly_trends, x='Month-Year', y='Quantity', marker='o', color='green')
plt.title('Monthly Order Quantity Trends')
plt.xlabel('Month-Year')
plt.ylabel('Total Quantity')
plt.xticks(rotation=45)
plt.show()
```

# **Business Insights**

```python
# Convert 'Order Date' to datetime format
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')

# Identify the Most Profitable Products and Categories
# Group by Product to find total profit and sales
product_profit = df.groupby('Product').agg({'Profit': 'sum', 'Sales': 'sum'}).reset_index()
top_products = product_profit.sort_values(by='Profit', ascending=False).head(10)  # Top 10 products by profit

# Group by Category to find total profit
category_profit = df.groupby('Category').agg({'Profit': 'sum', 'Sales': 'sum'}).reset_index()
top_categories = category_profit.sort_values(by='Profit', ascending=False)  # All categories sorted by profit

#Determine the Effect of Discounts on Sales and Profits
# Create a scatter plot to visualize the effect of discounts
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Discount', y='Sales', label='Sales', color='blue')
sns.scatterplot(data=df, x='Discount', y='Profit', label='Profit', color='orange')
plt.title('Effect of Discounts on Sales and Profit')
plt.xlabel('Discount')
plt.ylabel('Amount ($)')
plt.legend()
plt.show()

# Analyze Customer Behavior Based on Segmentation
# Group by Segment to find total profit
segment_profit = df.groupby('Segment').agg({'Profit': 'sum', 'Sales': 'sum', 'Customer ID': 'nunique'}).reset_index()
top_segment = segment_profit.sort_values(by='Profit', ascending=False)  # Sort segments by profit

# Discover Most Profitable and Underperforming Regions
# Group by Region to find total profit
region_profit = df.groupby('Region').agg({'Profit': 'sum', 'Sales': 'sum'}).reset_index()
region_profit = region_profit.sort_values(by='Profit', ascending=False)  # Sort by profit

# Investigate Which Countries Have the Highest Sales Volume
# Group by Country to find total sales volume
country_sales = df.groupby('Country').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()
country_sales = country_sales.sort_values(by='Sales', ascending=False)  # Sort by sales

# Display insights
print("Most Profitable Products:\n", top_products)
print("\nMost Profitable Categories:\n", top_categories)
print("\nEffect of Discounts on Sales and Profits: See the scatter plot above.")
print("\nCustomer Behavior by Segment:\n", top_segment)
print("\nProfit by Region:\n", region_profit)
print("\nSales Volume by Country:\n", country_sales)
```
